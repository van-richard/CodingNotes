The code looks like:

```
crossfade 48
hide #1 ribbon,atom
wait 48
```

One liner:

```
crossfade 48; hide #1 ribbon,atom; wait 48
```

Where:

crossfade [frames] - fade for # of frames
[action] - can be hiding ribbon, show atoms, color, etc.
wait [frames] - wait this number of frames before executing next line
